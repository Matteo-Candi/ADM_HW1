Alphabet Rangoli

```python
import string

def print_rangoli(size):
    alph = string.ascii_lowercase
    letters = alph[:size]
    w = (size-1) * 4 +1
    s = ""
    count = 1
    for i in range(size-1, -1, -1):
        line = []
        line.append(letters[i])
        for j in range(1, count):
            line.insert(0, letters[i+j])
            line.append(letters[i+j])
        count += 1
        s = s + "-".join(line).center(w, "-") + "\n"
    
    count = size - 1
    for i in range(1, size):
        line = []
        line.append(letters[i])
        for j in range(1, count):
            line.insert(0, letters[i+j])
            line.append(letters[i+j])
        count -= 1
        s = s + "-".join(line).center(w, "-") + "\n"
    print(s)
```

The Minion Game

```python
import string

def minion_game(s):
    vowels = ['A', 'E', 'I', 'O', 'U']
    alphabet = string.ascii_uppercase
    consonants = [x for x in alphabet if x not in vowels]
    
    count_stuart = 0
    for pos, el in enumerate(s):
        if el in consonants:
            count_stuart += len(s) - pos
    
    count_kevin = 0
    for pos, el in enumerate(s):
        if el in vowels:
            count_kevin += len(s) - pos
              
    if count_stuart > count_kevin:
        print(f"Stuart {count_stuart}")
    elif count_stuart < count_kevin:
        print(f"Kevin {count_kevin}")
    else:
        print("Draw") 
```

Piling Up!

```python
from collections import deque

def pile(n, array):
    d = deque(array)
    last = min(d)
    pile = max(d)
    for i in range(n):
        if d[-1] >= d[0]:
            if d[-1] <= pile:
                pile = d[-1]
                d.pop()
            else:
                print("No")
                break
        else:
            if d[0] <= pile:
                pile = d[0]
                d.popleft()
            else:
                print("No")
                break
    if pile == last:
        print("Yes")

T = int(input())
for i in range(T):
    n = int(input())
    A = list(map(int, input().split()))
    pile(n, A)
```

```python
import math
import os
import random
import re
import sys
from datetime import datetime


def time_delta(t1, t2):
    tz1 = datetime.strptime(t1, '%a %d %b %Y %H:%M:%S %z')
    tz2 = datetime.strptime(t2, '%a %d %b %Y %H:%M:%S %z')
    return str(abs(int((tz1 - tz2).total_seconds())))

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input())

    for t_itr in range(t):
        t1 = input()

        t2 = input()

        delta = time_delta(t1, t2)

        fptr.write(delta + '\n')

    fptr.close()
```

HTML Parser - Part 1 

```python
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print("Start :", tag)
        if len(attrs) != 0:
            for att in attrs:
                print(f"-> {att[0]} > {att[1]}")

    def handle_endtag(self, tag):
        print("End   :", tag)
        
    def handle_startendtag(self, tag, attrs):
        print("Empty :", tag)
        if len(attrs) != 0:
            for att in attrs:
                print(f"-> {att[0]} > {att[1]}")
    
parser = MyHTMLParser()
active = True
for _ in range(int(input())):
    line = input()
    if line[:3] == "<!--":
        active = False
    if line[-3:] == "-->":
        active = True
    if active:
        parser.feed(line)
```

```python
import re

for _ in range(int(input())):
    line = input()
    if len(line) == 10 and len(set(line)) == len(line) and line.isalnum():
        n_upper = len(re.findall(r'[A-Z]', line))
        n_number = len(re.findall(r'[0-9]', line))
        if n_upper >= 2 and n_number >= 3:
            print("Valid")
        else:
            print("Invalid")
    else:
        print("Invalid")
```

Validating Credit Card Numbers

```python
import re

for _ in range(int(input())):
    card = input()
    if len(re.findall(r'\-', card)) == 3:
        s = card.split("-")
        card = "".join(s)
        for el in s:
            if len(el) != 4:
                card = "a"
                break
    if card.isnumeric() and len(re.findall(r'[0-9]',card)) == 16 and re.match(r'[4|5|6]', card) != None and re.findall(r'([0-9])\1{3,}', card) == [] :
        print("Valid")
    else:
        print("Invalid")
```

# XML

XML 1 - Find the Score

```python
import sys
import xml.etree.ElementTree as etree

def get_attr_number(node):
    count = 0
    count += len(node.attrib)
    for child in node:
        if child.tag == "entry":
            for c in child:
                count += len(c.attrib)
        else:
            count += len(child.attrib)
    return count

if __name__ == '__main__':
    sys.stdin.readline()
    xml = sys.stdin.read()
    tree = etree.ElementTree(etree.fromstring(xml))
    root = tree.getroot()
    print(get_attr_number(root))
```

XML2 - Find the Maximum Depth

```python
import xml.etree.ElementTree as etree

maxdepth = 0
def depth(node, level):
    global maxdepth
    if len(node) == 0:
        maxdepth = -level-1
    else:
        if -level > maxdepth:
            maxdepth = -level
        for el in node:
            if len(el) != 0:
                depth(el, level-1)


if __name__ == '__main__':
    n = int(input())
    xml = ""
    for i in range(n):
        xml =  xml + input() + "\n"
    tree = etree.ElementTree(etree.fromstring(xml))
    depth(tree.getroot(), -1)
    print(maxdepth)
```

# CLOSURES AND DECORATIONS

Standardize Mobile Number Using Decorators

```python
import re

def wrapper(f):
    def fun(l):
        for pos, el in enumerate(l):
            if len(el) > 10:
                if el[0] == "0":
                   el = re.sub('^0', r'+91 ', el) 
                elif el[0] == "+":
                    el = re.sub('^(\+\d{2})', r'\1 ', el)
                elif el[0] == "9":
                    el = re.sub('^(\d\d)', r'+\1 ', el)
            else:
                el = re.sub('^(\d)', r'+91 \1', el)
            
            el = re.sub('(\d{5})', r'\1 ', el)
            l[pos] = el
        f(l)
    return fun


@wrapper
def sort_phone(l):
    print(*sorted(l), sep='\n')

if __name__ == '__main__':
    l = [input() for _ in range(int(input()))]
    sort_phone(l) 
```

Decorators 2 - Name Directory

```python
import operator

def person_lister(f):
    def inner(people):
        people.sort(key = lambda x: int(x[2]))
        for pos, person in enumerate(people):
            people[pos] = f(person)
        return people
    return inner

@person_lister
def name_format(person):
    return ("Mr. " if person[3] == "M" else "Ms. ") + person[0] + " " + person[1]

if __name__ == '__main__':
    people = [input().split() for i in range(int(input()))]
    print(*name_format(people), sep='\n')
```
